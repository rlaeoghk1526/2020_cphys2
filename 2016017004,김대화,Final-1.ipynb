```python
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
%matplotlib inline
```

와인데이터를 불러와서 기본적인 sklearn set를 불러와주는 코드입니다.

```python
wine = load_wine()
print(wine.DESCR)
```

데이터에 대한 설명입니다

```python
X=wine['data']
y=wine['target']
```

각각 data와 target로 분류해준 코드입니다.

```python
X.shape, y.shape
```

X,y 데이터의 규모를 알아본 코드 입니다

```python
wine.target_names
```

클러스터링을 y포함 3개의 값까지 가능함을 보여줍니다.

```python
wine.feature_names
```

클러스터링이 가능한 데이터값의 종류들 입니다.

```python
Xpart=[X[:,0],X[:,1]]
Xpart
```

```python
type(Xpart)
```

전체 데이터값 X를 첫번째와 두번째 열의 값끼리만 묶어서 array로 만든 다음 Xpart라는 새로운 list를 만들어 줬습니다.

```python
plt.scatter(Xpart[0],Xpart[1],c=y)
```

Xpart의 첫번째 값, 두번때값 그리고 y에 대한 분포 표 입니다.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3)
data = np.transpose(Xpart)
kmeans.fit(data)
```

이를 clustering 하기 위해서 수업시간에 배운 KMeans 기법을 도입했습니다.

```python
plt.scatter(Xpart[0],Xpart[1],c=kmeans.labels_)
```

눈에 보이듯 3등분을 한 것이 보입니다.

이는 다른 값에 대해서 clustering해도 잘 됨을 보여줍니다.

```python
Xpart=[X[:,1],X[:,2]]
data = np.transpose(Xpart)
kmeans.fit(data)

plt.figure(figsize=(10,10))
plt.subplot(221)
plt.scatter(Xpart[0],Xpart[1],c=y)

plt.subplot(222)
plt.scatter(Xpart[0],Xpart[1],c=kmeans.labels_)
```

```python
Xpart=[X[:,3],X[:,5]]
data = np.transpose(Xpart)
kmeans.fit(data)

plt.figure(figsize=(10,10))
plt.subplot(221)
plt.scatter(Xpart[0],Xpart[1],c=y)

plt.subplot(222)
plt.scatter(Xpart[0],Xpart[1],c=kmeans.labels_)
```
