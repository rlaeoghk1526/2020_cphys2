# 1.Boston House Price

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression, Ridge, Lasso
import matplotlib.pyplot as plt
import numpy as np
```

```python
boston = load_boston()
```

```python
X=boston.data
y=boston.target
```

```python
reg=LinearRegression()
lasso_reg=Lasso(alpha=10)
reg.fit(X,y)
lasso_reg.fit(X,y)
```

```python
x=np.array([i[12] for i in X])   #13번째 데이터 값에 대해서 입력한 것이다.
plt.scatter(x,y)        
```

```python
x=x.reshape(1,-1)
y=y.reshape(1,-1)
```

```python
reg=LinearRegression()
lasso_reg=Lasso(alpha=10)
reg.fit(x,y)
lasso_reg.fit(x,y)
```

```python
plt.scatter(x,reg.predict(x))
```

```python
plt.scatter(x,lasso_reg.predict(x))
```

# 2.Diabetes (당뇨)

###### bmi 지수에 대해 회귀 해 보겟습니다.

```python
from sklearn.datasets import load_diabetes
```

```python
dia = load_diabetes()
```

```python
K=dia.data
l=dia.target
```

```python
k=np.array([i[2] for i in K])  #<--- bmi지수 값에 대해 저장합니다
```

```python
plt.scatter(k,l)
```

```python
reg=LinearRegression()
lasso_reg=Lasso(alpha=2)
ridge_reg=Ridge(alpha=600000)
reg.fit(K,l)
lasso_reg.fit(K,l)
ridge_reg.fit(K,l)
```

```python
plt.scatter(k,reg.predict(K));
```

```python
plt.scatter(k,lasso_reg.predict(K))
```

```python
plt.scatter(k,ridge_reg.predict(K))
```
